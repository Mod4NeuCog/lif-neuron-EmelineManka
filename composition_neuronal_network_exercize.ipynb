```python
import numpy as np
class neuron :
    def __init__(self):
        self.stomach=100   
        self.V_th=-65
        self.V_rest= -75
        self.tau_m= 5
        self.g_L= 10
        self.V_init= -75
        self.E_L= -75
        self.I= 200
        self.dt= 0.1
        self.V=np.zeros(1000)
        self.V[0]=self.V_init
        
    def updateMembranePotential(self):
        for k in range(1,1000):
            dV=(-(self.V[k-1]-self.V_rest)+self.I/self.g_L)/self.tau_m
            if self.V[k-1]>=self.V_th:
                self.V[k]=self.V_init
            else:
                self.V[k]=self.V[k-1]+(self.dt*dV)
    
    def becoming_hungry(self):
        self.stomach=self.stomach-50
        
    def get_fed(self,nutrients):
        self.stomach=self.stomach+nutrients
        
        
        
class glial_cell:
    
    def __init__(self):
        self.nutrient=200
        
    def get_meal(self):
        return self.nutrient
    
    
class Neuron_network :
    def __init__(self,n,g):
        self.neurons=[]
        self.glial_cells=[]
        for i in range(n):
            N=neuron()
            self.neurons.append(N)
        for j in range(g):
            G=glial_cell()
            self.glial_cells.append(G)
        
    def update_mb_of_all_neurons(self,n):
        for k in range(n):
            self.neurons[k].updateMembranePotential()
    
    
    
def main():
    my_network=Neuron_network(100,1000)
    my_network.update_mb_of_all_neurons(100)
    
    print(my_network.neurons[5].V)
    #print(my_network.glial_cells)
        
```
