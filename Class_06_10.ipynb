```python
    
def dVdt(t):
    last_V=Vlist[t-1]
    return (-(last_V-V_reset)+(I/g_L))/tau_m
    
def V(t):
    last_V=Vlist[t-1]
    return last_V+(dVdt(t)*dt)
```

nd here is the simulator :

```python
I=200 #basic parameters
V_reset=-75
V_th=-65
dt=0.1
tau_m=5
g_L=10
V_init=-75
Vlist=[] #List for potential values
Xlist=[]    #List for timeline
Xlist.append(0)     #adds the first time, at 0
Vlist.append(V_init)    #adds the first potential, V_init
    
```

) simulate neuron for t=0 to t=5

```python
def Simulator1(T_0,T):
    for t in range(T_0+1,T*100): #because 5 is too small
        Xlist.append(t)
        if V(t)>V_th:
            Vlist.append(V_reset)
        else:
            Vlist.append(V(t))
    x=np.array(Xlist)
    y=np.array(Vlist)
    plt.plot(x,y) 
    plt.show()
```

```python
def Simulator2(T):
    t=1
    while t<T*100 : #because 5 is too small
        Xlist.append(t)
        if V(t)>V_th:
            Vlist.append(V_reset)
        else:
            Vlist.append(V(t))
        t+=1
    x=np.array(Xlist)
    y=np.array(Vlist)
    plt.plot(x,y) 
    plt.show()
```
